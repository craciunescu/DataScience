```python
import numpy as np
data = [] 

#Function for reading file
def readData(file):
    openfile = open(file, "r")
    for linea in openfile:
        data.append(linea)
    return data

#Function for deleting unusefull character as tabs
def cleanData(data):
    listFilter = []
    for i in range(len(data)):
        fil = "".join(map(str, list(filter(lambda x: x!= '\t', list(data[i]))))).split()
        listFilter.append(fil)
    return listFilter

usefulData = cleanData(readData("cars.txt"))
showresult(usefulData, 0.4,0.9)

    
```

Before doing any analysis, it is necessary cleaning correctly the data:

```python
# Deleting spaces in the description 
data['Description'] = data['Description'].str.strip() 
```

The aim of this analysis is findind the between the items bougth by people from different countries.

```python
# Transactions done in Spain
Spain = (data[data['Country'] =="Spain"]
              .groupby(['InvoiceNo', 'Description'])['Quantity'] 
              .sum().unstack().reset_index().fillna(0) 
              .set_index('InvoiceNo')) 
```

```python
# Transforming data into boolean values that will be use before
def toBool(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
Spain = Spain.applymap(toBool)
```
